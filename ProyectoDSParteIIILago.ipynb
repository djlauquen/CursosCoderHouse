# Proyecto Data Science

**Abstract:** El objetivo de este trabajo consiste en analizar ciertas variables climáticas de la ciudad de Washington, en el intervalo temporal desde el año 2015 hasta el año 2024. Para ello se dispone de un dataset de 33 columnas y 3319 filas, donde cada fila contiene la información de las distintas variables climáticas concernientes a un solo día, y cada columna contiene la información diaria a través de los años de una dada variable climática. Para este trabajo, las variables climáticas de interés son la temperatura (en grados centígrados), la humedad relativa (en porcentaje) y las precipitaciones (en milímetros). El dataset en cuestión será convertido en un DataFrame mediante la librería Pandas.

Primero vamos a focalizarnos en encontrar algún tipo de correlación entre las variables citadas mediante el análisis de gráficos de dispersión, y luego vamos a estudiar la evolución temporal del promedio de las tres variables a lo largo de los años utilizando gráficos de barras. Para esta segunda parte, vamos a agregar una última columna al DataFrame que contemple solo el año, lo cual nos permitirá realizar los gráficos necesarios para el análisis. Los distintos gráficos tanto de dispersión como de barras se realizarán con las librerías Matplotlib y Pandas.

Finalmente, vamos a focalizarnos en estudiar las siguientes variables de interés: radiación solar, humedad, temperatura y precipitaciones medias. Calcularemos la matriz de correlación entre dichas variables y luego aplicaremos el modelo de regresión lineal a aquelas variables que presenten mayor correlación.

Fuente: https://www.kaggle.com/datasets/taweilo/washington-dc-historical-weather-20158202407

```python
# Importar las librerías necesarias usando sus respectivas alias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Leer el archivo CSV a analizar

df = pd.read_csv('dc_weather.csv')
df.head()
```

```python
# Mostrar información general del DataFrame

df.info()
```

```python
# Agregamos una última columna con el año al DataFrame

df['year'] = df['datetime'].str.slice(0, 4).astype(dtype = 'int')
```

```python
# Mostrar información general del DataFrame

df.info()
```

```python
# Descripción estadística de las columnas numéricas

df.describe()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['temp'],df['humidity'])
plt.xlabel('Temperatura (°C)')
plt.ylabel('Humedad (%)')
plt.title('Gráfico de Dispersión Temperatura vs Humedad')
plt.grid(True)
plt.show()
```

De este gráfico podemos concluir que no existe una correlación clara entre la temperatura y la humedad relativa ya que la dispersión de puntos es de un tamaño tal que no es posible distinguir algún patrón o comportamiento que relacione a ambas variables.

```python
plt.figure(figsize=(10,6))
plt.scatter(df['temp'],df['precip'])
plt.xlabel('Temperatura (°C)')
plt.ylabel('Precipitaciones (mm)')
plt.title('Gráfico de Dispersión Precipitaciones vs Temperatura')
plt.grid(True)
plt.show()
```

En este gráfico se puede apreciar un leve aumento de las precipitaciones promedio a medida que aumenta la temperatura.

```python
plt.figure(figsize=(10,6))
plt.scatter(df['precip'],df['humidity'])
plt.xlabel('Precipitaciones (mm)')
plt.ylabel('Humedad (%)')
plt.title('Gráfico de Dispersión Humedad vs Precipitaciones')
plt.grid(True)
plt.show()
```

A partir de este gráfico queda bastante claro que existe una clara relación entre la humedad y las precipitaciones. En efecto, a medida que aumenta la humedad relativa, no solo se tienen bajas precipitaciones, sino que además aumentan las altas precipitaciones.

## Análisis de la correlación de las variables de interés

Vamos a analizar la correlación de las siguientes variables de interés: temperatura, humedad relativa, precipitaciones y radiación solar.

Para ello, primero vamos a crear un segundo DataFrame que solo contenga estas cuatro columnas del DataFrame original.

```python
columnas_seleccionadas = ["temp", "humidity", "precip", "solarradiation"]
df2 = df[columnas_seleccionadas]
print(df2)
```

A continuación, vamos a calcular la matriz de correlación entre estas variables de interés, y la vamos a visualizar.

```python
correlation_matrix = df2.corr()
print(correlation_matrix)
# Mostrar el gráfico de correlación
sns.heatmap(correlation_matrix, annot = True)
plt.show()
```

Podemos apreciar de la matriz de correlación que las variables que presentan mayor correlación son la radiación solar con la temperatura y la radiación solar con la humedad. Más aun, vemos que la relación entre la radiación solar y la temperatura es directa (ambas crecen o ambas decrecen) ya que presentan un número de correlación positiva, mientras que la relación entre la radiación solar y la humedad es inversa (cuando una variable crece, la otra decrece) pues presentan un número de correlación negativa.

Finalmente, vamos a aplicar el modelo de regresión lineal para la humedad en función de la radiación solar y para la temperatura también en función de la radiación solar.

# Modelo de Regresión Lineal para las variables de interés

### Modelo de regresión lineal para la humedad relativa en función de la radiación solar

```python
# Conjunto de datos de entrenamiento y prueba

X = df[['solarradiation']]
y = df['humidity']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# Entrenamiento del modelo de regresión lineal

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predicción de la humedad relativa en función de la radiación solar

y_pred = model.predict(X_test)
```

```python
# Evaluación del rendimiento del modelo 

print("Mean Squared Error:", mean_squared_error(y_test, y_pred))  
print("R^2 Score:", r2_score(y_test, y_pred))
```

```python
# Visualización del modelo

plt.figure(figsize=(10,5))
plt.scatter(X_test, y_test, color = 'gray', label = 'Datos Reales')
plt.plot(X_test, y_pred, color = 'red', linewidth = 2, label = 'Predicciones')
plt.xlabel('Radiación Solar (W/m2)')
plt.ylabel('Humedad Relativa (%)')
plt.title('Regresión Lineal: Radiación Solar vs Humedad Relativa')
plt.legend()
plt.show()
```

Vemos que la recta asociada a la humedad relativa en función de radiación solar tiene pendiente negativa, consistente con el valor negativo de correlación entre ambas variables obtenido anteriormente.

### Modelo de regresión lineal para la temperatura en función de la radiación solar

```python
# Conjunto de datos de entrenamiento y prueba

X = df[['solarradiation']]
y = df['temp']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# Entrenamiento del modelo de regresión lineal

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predicción de la temperatura en función de la radiación solar

y_pred = model.predict(X_test)
```

```python
# Evaluación del rendimiento del modelo 

print("Mean Squared Error:", mean_squared_error(y_test, y_pred))  
print("R^2 Score:", r2_score(y_test, y_pred))
```

```python
# Visualización del modelo

plt.figure(figsize=(10,5))
plt.scatter(X_test, y_test, color = 'gray', label = 'Datos Reales')
plt.plot(X_test, y_pred, color = 'red', linewidth = 2, label = 'Predicciones')
plt.xlabel('Radiación Solar (W/m2)')
plt.ylabel('Temperatura (°C)')
plt.title('Regresión Lineal: Radiación Solar vs Temperatura')
plt.legend()
plt.show()
```

Vemos que la recta asociada a la temperatura en función de radiación solar tiene pendiente positiva, consistente con el valor positivo de correlación entre ambas variables obtenido anteriormente.

Notar de ambos gráficos dos puntos bastante a la derecha del cúmulo de puntos que se encuentra mas bien a la izquierda; estos dos puntos son posiblemente dos outliers que, si se hubiesen tenido en cuenta, posiblemente el modelo de regresión lineal hubiese dado mejores resultados.
